```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

**1. Load dataset**

```python

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

**2. Preprocessing**

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

**Reshape for CNN (28x28x1)**

```python
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

**Model (CNN)**

```python

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

**Compile**

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

**Train**

```python
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

**Accuracy**

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_acc)
```

```python
plt.imshow(x_test[0].reshape(28,28), cmap='gray')
plt.title(f"Predicted: {model.predict(x_test[:1]).argmax()}")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

for i in range(6):
    pred = model.predict(x_test[i:i+1])
    predicted_label = np.argmax(pred, axis=1)[0]

    plt.subplot(2,3,i+1)
    plt.imshow(x_test[i].reshape(28,28), cmap='gray')
    plt.title(f"P:{predicted_label} A:{y_test[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

wrong = []

for i in range(len(x_test)):
    pred = np.argmax(model.predict(x_test[i:i+1]), axis=1)[0]
    if pred != y_test[i]:
        wrong.append((i, pred, y_test[i]))
    if len(wrong) == 5:
        break

plt.figure(figsize=(10,5))

for i, (idx, pred, actual) in enumerate(wrong):
    plt.subplot(1,5,i+1)
    plt.imshow(x_test[idx].reshape(28,28), cmap='gray')
    plt.title(f"P:{pred} A:{actual}")
    plt.axis('off')

plt.show()
```
